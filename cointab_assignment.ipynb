```python
#remaing columns name's to easy understand
order_data.rename(columns={'Order Qty':'Order_Qty'},inplace = True)
order_data.rename(columns={'Weight (g)':'Weight'},inplace = True)
order_data.rename(columns={'Item Price(Per Qty.)':'Item_Price'},inplace = True)
order_data.rename(columns={'Billing Amount (Rs.)':'Billing_Amount'},inplace = True)
order_data.rename(columns={'Payment Mode':'Payment_Mode'},inplace = True)
```

```python
# Calculate total weight per order based on SKU quantity and gross weight
order_data['Total Weight'] = order_data['Order_Qty'] * order_data['Weight']
```

```python
# Calculate expected COD charges
def calculate_cod_charge(payment_Mode, Billing_Amount):
    if payment_Mode == 'COD':
        if Billing_Amount  <= 300:
            return 15
        else:
            return 0.05 * Billing_Amount
    else:
        return 0
order_data['Expected COD Charge'] = order_data.apply(lambda row: calculate_cod_charge(row['Payment_Mode'], row['Billing_Amount']), axis=1)
```

```python
def calculate_applicable_weight_slab(Weight, Zone):
    slab_count = order_data[order_data['Zone_x'] == Zone]['Weight Slabs'].values[0]
    return (slab_count + 1) * slab_count
# Calculate the applicable weight slab for each order and add to the order_data DataFrame
order_data['Applicable Weight Slab'] = order_data.apply(lambda row: calculate_applicable_weight_slab(row['Total Weight'], row['Zone_x']), axis=1)
```

```python
order_data
```

```python
from pandas.core.dtypes.dtypes import Ordered
# Calculate total forward charge based on weight and zone
def calculate_forward_charge(Weight, Zone):
    rate_row = order_data[(order_data['Zone_x'] == Zone) & (order_data['Forward Fixed Charge'] == 'fwd') & (order_data['Weight Slabs'] == calculate_applicable_weight_slab(Weight, Zone))]
    return order_data['Forward Fixed Charge'].values[0] + (order_data['Forward Additional Weight Slab Charge'].values[0] * (Weight - order_data['Weight Slabs'].values[0]))


order_data['Total Forward Charge'] = order_data.apply(lambda row: calculate_forward_charge(row['Total Weight'], row['Zone_x']), axis=1)
```

```python
# Calculate total RTO charge based on weight and zone
def calculate_rto_charge(Weight, Zone):
    rate_row = order_data[(order_data['Zone_x'] == Zone) & (order_data['RTO Fixed Charge'] == 'rto') & (order_data['Weight Slabs'] == calculate_applicable_weight_slab(Weight, Zone))]
    return order_data['RTO Fixed Charge'].values[0] + (order_data['RTO Additional Weight Slab Charge'].values[0] * (Weight - order_data['Weight Slabs'].values[0]))

order_data['Total RTO Charge'] = order_data.apply(lambda row: calculate_rto_charge(row['Total Weight'], row['Zone_x']), axis=1)
```

```python
# Calculate total charges
order_data['Total Charges'] = order_data['Total Forward Charge'] + order_data['Total RTO Charge'] + order_data['Expected COD Charge']
```

```python
# Calculate the expected charges for each order
order_data['Expected Charges'] = (
    order_data['Total Forward Charge'] + order_data['Total RTO Charge'] + order_data['Difference']
)

# Compare expected charges with billed charges
order_data['Charge Difference'] = order_data['Billing_Amount'] - order_data['Expected Charges']

# Categorize orders based on charge difference
order_data['Charge Category'] = pd.cut(
    order_data['Charge Difference'],
    bins=[float('-inf'), -0.01, 0.01, float('inf')],
    labels=['Undercharged', 'Correctly Charged', 'Overcharged']
)

# Create the summary table
summary_table = order_data.groupby('Charge Category').agg(
    Count=pd.NamedAgg(column='Order ID', aggfunc='count'),
    Amount=pd.NamedAgg(column='Billing_Amount', aggfunc='sum')
)

# Display the summary table
print(summary_table)
```
